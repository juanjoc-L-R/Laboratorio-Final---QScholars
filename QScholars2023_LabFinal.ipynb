```python
from qiskit import QuantumCircuit, Aer, assemble
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import partial_trace

# Crear el circuito cuántico
qc = QuantumCircuit(3)

# Aplicar las puertas necesarias para generar el estado |ψ⟩
qc.h(0)
qc.h(1)
qc.x(2)
qc.h(2)

# Obtener el estado cuántico resultante
simulator = Aer.get_backend('statevector_simulator')
result = simulator.run(assemble(qc)).result()
statevector = result.get_statevector()

# Calcular la matriz de densidad reducida para el segundo qubit
rho_partial = partial_trace(statevector, [0, 2])

# Obtener las probabilidades de los estados |0⟩ y |1⟩ para el qubit central
prob_0 = rho_partial.data[0][0]
prob_1 = rho_partial.data[1][1]

print("Probabilidad de |0⟩:", prob_0)
print("Probabilidad de |1⟩:", prob_1)
```

```python
#initialization
import matplotlib.pyplot as plt
import numpy as np
import math

# importing Qiskit
from qiskit import IBMQ, Aer, transpile, execute
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
#from qiskit.providers.ibmq import least_busy

# import basic plot tools
from qiskit.visualization import plot_histogram
```

```python
#We start by preparing a quantum circuit with 3 qubits:
n = 3
grover_circuit = QuantumCircuit(n)
```

```python
qc = QuantumCircuit(3)
# Apply CZ gates to mark the states |011>, |010>, |100>, and |101>
qc.cz(0, 2)  # Marks |011>
qc.cz(1, 2)  # Marks |010>
qc.cz(0, 2)  # Marks |100> (Repeat the CZ gate on qubit 0)
qc.cz(0, 2)  # Marks |101> (Repeat the CZ gate on qubit 0)

# Create a gate from the circuit
oracle_ex3 = qc.to_gate()
oracle_ex3.name = "U$_\omega$"
```

```python
def diffuser(nqubits):
    qc = QuantumCircuit(nqubits)
    # Apply transformation |s> -> |00..0> (H-gates)
    for qubit in range(nqubits):
        qc.h(qubit)
    # Apply transformation |00..0> -> |11..1> (X-gates)
    for qubit in range(nqubits):
        qc.x(qubit)
    # Do multi-controlled-Z gate
    qc.h(nqubits-1)
    qc.mct(list(range(nqubits-1)), nqubits-1)  # multi-controlled-toffoli
    qc.h(nqubits-1)
    # Apply transformation |11..1> -> |00..0>
    for qubit in range(nqubits):
        qc.x(qubit)
    # Apply transformation |00..0> -> |s>
    for qubit in range(nqubits):
        qc.h(qubit)
    # We will return the diffuser as a gate
    U_s = qc.to_gate()
    U_s.name = "U$_s$"
    return U_s
```

```python
#First, we need to initialize the state . Let's create a general function (for any number of qubits) so we can use it again later:
def initialize_s(qc, qubits):
    """Apply a H-gate to 'qubits' in qc"""
    for q in qubits:
        qc.h(q)
    return qc
```

```python
n = 3
grover_circuit = QuantumCircuit(n)
grover_circuit = initialize_s(grover_circuit, [0,1,2])
grover_circuit.append(oracle_ex3, [0,1,2])
grover_circuit.append(diffuser(n), [0,1,2])
grover_circuit.measure_all()
grover_circuit.draw()
```

#problema 6


En un oráculo balanceado en el contexto de algoritmos cuánticos como Grover's, la compuerta controlada-X (CX o CNOT) se utiliza para aplicar una operación lógica XOR entre el estado del qubit de control y el estado del qubit objetivo. Esta operación lógica es esencial para lograr la balanza entre las soluciones y no soluciones, lo que es crucial para el funcionamiento de algoritmos como Grover's.

En términos generales, en un oráculo balanceado, no puedes prescindir de ningún qubit de control en la compuerta CX sin afectar la balance de soluciones y no soluciones. Cada qubit de control en la compuerta CX contribuye a definir la función booleana que el oráculo representa.

#problema 7

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.circuit.library import XGate, ZGate
from qiskit.visualization import plot_histogram
from qiskit import execute, Aer

def oracle_part(qc, qr):
    mcx = XGate().control(3)

    # Implementa las aristas del grafo
    qc.cx(qr[0], qr[4])
    qc.cx(qr[0], qr[5])
    qc.cx(qr[1], qr[6])
    qc.cx(qr[1], qr[7])
    qc.cx(qr[2], qr[8])
    qc.cx(qr[2], qr[9])
    qc.cx(qr[3], qr[10])
    qc.cx(qr[3], qr[11])

     #Los qubits del 9 al 11 nos van a permitir contar la cantidad de aristas correctas
    #de tal forma que la suma será expresada en binario con cifras dadas por q_11, q_10 y q_9.
    qc.barrier()
    qc.cx(qr[4],qr[9])
    qc.barrier()

    qc.ccx(qr[5], qr[9], qr[10])
    qc.cx(qr[5],qr[9])
    qc.barrier()

    qc.ccx(qr[6], qr[9], qr[10])
    qc.cx(qr[6],qr[9])
    qc.barrier()

    qc.append(mcx,[7,9,10,11])
    qc.ccx(qr[7], qr[9], qr[10])
    qc.cx(qr[7],qr[9])
    qc.barrier()

    qc.append(mcx,[8,9,10,11])
    qc.ccx(qr[8], qr[9], qr[10])
    qc.cx(qr[8],qr[9])
    qc.barrier()

    qc.x(qr[9])
    qc.x(qr[10])
    qc.barrier()

    #El qubit 12 indicará cuando se haya encontrado un estado deseado
    qc.append(mcx,[9,10,11,12])


    return qc

def oracle(qc, qr):
    qc = oracle_part(qc, qr)
    qc1 = qc.inverse()
    qc.z(12)
    qc.compose(qc1, inplace=True)

    return qc

def grover_diffusion(qc, qr):
    mcz = ZGate().control(3)

    qc.barrier()
    qc.h(range(4))
    qc.x(range(4))
    qc.append(mcz, list(range(0, 4)))
    qc.x(range(4))
    qc.h(range(4))
    qc.barrier()

    return qc

def grover(qc, qr):
    return grover_diffusion(oracle(qc, qr), qr)

# Crear los registros cuánticos
qr = QuantumRegister(13)
qra = QuantumRegister(13)
cra = ClassicalRegister(4)

#definiendo qc
qc=QuantumCircuit(qr)

# Crear los circuitos cuánticos
qa = QuantumCircuit(qra, cra)
qa.h(range(4))
qb = grover(qc, qr)
qa.compose(qb.compose(qb), inplace=True)
qa.measure(range(4), cra)

# Ejecutar la simulación
qasm_simulator = Aer.get_backend('qasm_simulator')
job = execute(qa, qasm_simulator, shots=1000)
counts = job.result().get_counts(qa)

plot_histogram(counts)
```

```python
from qiskit import QuantumCircuit, transpile, assemble, Aer, execute
import numpy as np
import math

def qft(circuit, n):
    for j in range(n):
        circuit.h(j)
        for k in range(j+1, n):
            circuit.cp(2*np.pi/2**(k-j+1), k, j)
        circuit.barrier()

# Crear el circuito cuántico
n = 4  # Número de qubits
qft_circuit = QuantumCircuit(n)
qft(qft_circuit, n)

# Simulador unitario de Qiskit
simulator = Aer.get_backend('unitary_simulator')

# Ejecutar el circuito en el simulador unitario
job = execute(qft_circuit, simulator)
result = job.result()
unitary_matrix = result.get_unitary()

# Imprimir la matriz unitaria resultante
print("Matriz unitaria de la Transformada Cuántica de Fourier:")
print(unitary_matrix)

# Imprimir el circuito cuántico
print("\nCircuito cuántico de la Transformada Cuántica de Fourier:")
print(qft_circuit)
```
